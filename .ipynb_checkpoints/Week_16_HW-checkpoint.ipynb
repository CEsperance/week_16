**1. Perform combined over and undersampling on the diabetes dataset (use SMOTEENN). Explain how combined sampling works.**

In general, resampling methods are designed to add or remove examples from the training dataset in order to change the class distribution. SMOTE-ENN (SMOTE and Edited Nearest Neighbors Undersampling) is a combined resampling method to balance an imbalanced dataset by adding oversampling the minority class and undersampling the majority class. 

SMOTE is the oversampling technique used in combination with Edited NearestNeighbors(ENN), which is an undersampling technique. ENN involves using k=3 nearest neighbors to locate examples in a dataset that are misclassified and then removed from both classes in the training set. It can be applied to all classes or just those examples in the majority class. Below the sampling strategy for ENN is set to all, so it removes misclassified data from both classes in this dataset.

The class size for those without diabetes has been decreased to 392, while the class of those who do have diabetes has been increased to the same sample size.

```python
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import EditedNearestNeighbours
from collections import Counter

#define and combine SMOTE and Edited Nearest Neighbors resampling for imbalanced classification

sme=SMOTEENN(random_state=24, enn=EditedNearestNeighbours(sampling_strategy='all'))

X_resampled, y_resampled = sme.fit_resample(X_train_scaler, y_train)
print('Resampled dataset shape %s' % Counter(y_resample))
```

a.	BRIEFLY explain what the ROC Curve shows and what the resulting AUC means. 

The ROC (Receiver Operating Characteristic) curve is used to predict the probablity of a binary outcome.
- summarize the trade-off between the true positive rate and false positive rate for a predictive model using different probability thresholds.
- It is a plot of the false positive rate (x-axis) versus the true positive rate (y-axis) for a number of different candidate threshold values between 0.0 and 1.0.

The shape of the curve shows that the smaller values on the x-axis of the plot indicate lower false positives and higher true negatives. Larger values on the y-axis of the plot indicate higher true positives and lower false negatives.

The area under the curve shows that this model is able to capture 82% of the true positive diabetes cases will be classified correctly.



#helpful link to reference https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/

**4. What is outlier detection? Why is it useful? What methods can you use for outlier detection?**

Outlier detection analyzes data points whose feature values are different from the normal points in a dataset. Detecting outliers is useful because it finds bad data or data that is noisy, which can have a negative influence on predictive models.

Methods to used for outlier detection depends on the source of the data that needs to be analyzed and the homogeneity of the cource of data. Some methods are: 

- KNN:a proximity based technique. The farther a data point is beyond its neighbors, the more possible the data is an outlier
- Dbscan (Density Based Spatial Clustering of Applications with Noise): a clustering method to detect outliers in nonparametric distributions in many dimensions.
- Z-Score (parametric): a metric that indicates how many standard deviations a data point is from the samples mean, assuming a gaussian distribution
- Linear Regression and Statsmodel's OLS : the r squared value will inform you if the variance in the dataset is explained by the predictors that were chosen. You want a high r-sqaured value.

**5.Perform a linear SVM to predict credit approval (last column) using this dataset: https://archive.ics.uci.edu/ml/datasetsStatlog+%28Australian+Credit+Approval%29.  Make sure you look at the accompanying document that describes the data in the dat file. You will need to either convert this data to another file type or import the dat file to python.**

**a. How did the SVM model perform?**

A precision score of 90% for true negative values shows that out of the values that were predicted as negatives by the model, 90% were true negatives

A recall score of 79% for true negatives indicates that the model correctly predicted 79% true negatives. The model was also 89% correct in predicting true positives.

```python
#load the .dat file as an array
australia_array = sed = np.loadtxt('australian.dat')
australia_array
#Convert the array to a dataframe
credit_df = pd.DataFrame(australia_array)
credit_df
#new column names assigned
credit_df.columns = ["A1", "A2", "A3", "A4", "A5", "A6", "A7", "A8", "A9", "A10", "A11", "A12", "A13", "A14", "A15"]
#check new names
credit_df
```

```python
#checking for missing values
credit_df.isnull().sum()
```

```python
#split data to target variable and features

#features
X1 = credit_df.drop('A15', axis=1)
#target variable
y1 = credit_df['A15']

#train, test, split data
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.2, random_state = 24, stratify = y1 )

#Standardize
sc = StandardScaler()
X1_train = sc.fit_transform(X1_train)
X1_test = sc.fit_transform(X1_test)
```

```python
#https://www.kaggle.com/baiazid/pima-indians-diabetes-svm

#implement linear SVM
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
#fit the classifier
classifier.fit(X1_train, y1_train)
```

```python
#printing the training and validation accuracy.
print(classifier.score(X1_train, y1_train))
print(classifier.score(X1_test, y1_test))
```

```python
#predict test set results
y_pred = classifier.predict(X1_test)
```

```python
#Evaluate the model

from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y1_test, y_pred)
print (cm)
print(accuracy_score(y1_test, y_pred))
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(y1_test, y_pred))
```
